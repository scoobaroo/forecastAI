```python
uniqueDevices = df.Device.unique()
devicesAndQty = pd.DataFrame(df.groupby('Device')["Quantity"].sum())
devicesAndQty['Quantity'].sum()
```

```python
devices = df.groupby('Device')['Quantity'].sum()
devicesAndQts = pd.DataFrame({'Device':devices.index, 'Quantity':devices.values})
sortedDevices = devicesAndQts.sort_values("Quantity", ascending=False).head(10)
len(sortedDevices)
sortedDevices
```

```python
sortedDeviceList = sortedDevices["Device"].tolist()
devicesAndPartNumbers = {}
for device in sortedDeviceList:
    specificDevice = df.loc[df["Device"]==device]
    partNumbers = specificDevice['Part Number'].unique()
    devicesAndPartNumbers[device] = partNumbers.tolist()

len(devicesAndPartNumbers)
devicesAndPartNumbers
```

```python
import json
with open('sortedDevicesAndPartNumbers.json', 'w') as fp:
    json.dump(devicesAndPartNumbers, fp)
```

```python
opp["month&year"] = opp["Created On"].map(lambda x: str(x.month) +"/"+ str(x.year))
opp.sort_index(inplace=True)
monthAndyear = opp["month&year"].unique()
monthAndyear
endcustomer = opp["End Customer"].unique()
trainX = []
stringifiedCustomers = []
for c in endcustomer:
    stringifiedCustomer = str(c)
    stringifiedCustomers.append(stringifiedCustomer)
le.fit(stringifiedCustomers)

for mAndy in monthAndyear:
    for customer in endcustomer:
        num = le.transform([customer])
        feature = mAndy + "-" + str(num)
        trainX.append(feature)
```

```python
def listToStringWithoutBrackets(list1):
    return str(list1).replace('[','').replace(']','')
trainX = listToStringWithoutBrackets(trainX)
trainX
```

```python
train = pd.DataFrame()
for element in trainX:
    train["key"] = element
    
train["key"] = ""
train["quantity"] = 0
train
```
